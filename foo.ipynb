Checks whether x is divisible by k.

```python
def is_divisible_by_k(x,k):
  if(x%k==0):
    print("Divisible")
  else:
    print("Not divisible")
```

```python
# test the function:
is_divisible_by_k(10,2) # Should return "Divisible"
is_divisible_by_k(37,2) # Should return "Not divisible"
```

Store all the integers that are multiples of 2 or 5 or 7 that are lower or equal to 1000 (excluding doubles)

```python
x = [x for x in range(1000) if x%2 == 0 or x%5 == 0 or x%7 == 0]
```
